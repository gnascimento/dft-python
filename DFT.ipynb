```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def gen_wave(freq, amp, T, shift, sr):
    time = np.arange(0, T, T/sr)
    X = amp * np.sin(2*np.pi*freq*time+shift)
    return time, X
```

```python
N = 100
```

```python
time, amplitude = gen_wave(2, 3, 1, 0, N)
```

```python
_, amplitude2 = gen_wave(5, 2, 1, 0, N)
```

```python
amplitude3 = amplitude + amplitude2
```

```python
plt.figure(figsize=[15, 5])
plt.ylim([-5, 5])
plt.grid(True, which='both')
plt.plot(time, amplitude, c="b")
```

```python
plt.figure(figsize=[15, 5])
plt.ylim([-5, 5])
plt.grid(True, which='both')
plt.plot(time, amplitude2, c="b")
```

```python
plt.figure(figsize=[15, 5])
plt.ylim([-5, 5])
plt.grid(True, which='both')
plt.plot(time, amplitude3, c="b")
```

```python
#vetor indice
k = np.arange(0, N)
n = np.arange(0, N)
```

```python
k = k.reshape((N,1))
```

```python
M = 2 * np.pi * k * n / N
Xa = np.cos(M)
Xb = 1j * np.sin(M)
X = Xa - Xb
print(X.shape)
Y = X.dot(amplitude3)
print(Y.shape)
```

```python
plt.plot(k, Y.real, k, Y.imag)
```
